```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
import math
import collections
from scipy.spatial.distance import cityblock as cityblock
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
def draw_adjacency_matrix(G, node_order=None, partitions=[], colors=[]):
    """
    - G is a netorkx graph
    - node_order (optional) is a list of nodes, where each node in G
          appears exactly once
    - partitions is a list of node lists, where each node in G appears
          in exactly one node list
    - colors is a list of strings indicating what color each
          partition should be
    If partitions is specified, the same number of colors needs to be
    specified.
    """
    adjacency_matrix = nx.to_numpy_matrix(G, dtype=np.bool, nodelist=node_order)

    #Plot adjacency matrix in toned-down black and white
    fig = plt.figure(figsize=(5, 5)) # in inches
    plt.imshow(adjacency_matrix,
                  cmap="Greys",
                  interpolation="none")
    
    # The rest is just if you have sorted nodes by a partition and want to
    # highlight the module boundaries
    assert len(partitions) == len(colors)
    ax = plt.gca()
    for partition, color in zip(partitions, colors):
        current_idx = 0
        for module in partition:
            ax.add_patch(patches.Rectangle((current_idx, current_idx),
                                          len(module), # Width
                                          len(module), # Height
                                          facecolor="none",
                                          edgecolor=color,
                                          linewidth="1"))
            current_idx += len(module)
```

```python
a = nx.barabasi_albert_graph(21, 2)
b = nx.barabasi_albert_graph(21, 2)
c = nx.barabasi_albert_graph(31, 2)
d = nx.barabasi_albert_graph(41, 2)
graphs = [a,b,c,d]
```

```python
def changeEdge(g, e):
    x, y = e
    k = np.random.randint(len(g.nodes()))
    while(k==x or (x,k) in g.edges()):
        k = np.random.randint(len(g.nodes()))
        
    g.remove_edge(x,y)
    g.add_edge(x,k)
    return g

f = nx.Graph()
for g in graphs:
    idx = len(f.nodes())
    newNodes = [x+idx for x in g.nodes()]
    newEdges = [(x+idx,y+idx) for (x,y) in g.edges()]
    f.add_nodes_from(newNodes)
    f.add_edges_from(newEdges)

beta = 0.15
support_f = f.copy()
for e in support_f.edges():
    if beta > np.random.uniform():
        f = changeEdge(f,e)
```

```python
def hierarchicalGraph(n, k, beta, nOfCommunities):
	div = getDivArray(n,nOfCommunities)
	graphs = []
	for x in div:
		graphs.append(nx.barabasi_albert_graph(x, k))
		
	f = nx.Graph()
	for g in graphs:
		idx = len(f.nodes())
		newNodes = [x+idx for x in g.nodes()]
		newEdges = [(x+idx,y+idx) for (x,y) in g.edges()]
		f.add_nodes_from(newNodes)
		f.add_edges_from(newEdges)
		
	support_f = f.copy()
	for e in support_f.edges():
		if beta > np.random.uniform():
			f = changeEdge(f,e)	
	
	return f
	
def getDivArray(n,k):
	div = n/k
	asd = [div] * k
	if n%k != 0:
		asd[-1] = asd[-1] + (n%k)

	return asd
		
def changeEdge(g, e):
    x, y = e
    k = np.random.randint(len(g.nodes()))
    while(k==x or (x,k) in g.edges()):
        k = np.random.randint(len(g.nodes()))
        
    g.remove_edge(x,y)
    g.add_edge(x,k)
    return g
```
