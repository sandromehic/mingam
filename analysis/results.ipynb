```python
# import neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from os import listdir
import itertools
%matplotlib inline

import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
def readFullGamedata(dirname):
    # read all the gamedata files from a given folder
    # output - an dictionary of gamedata
    
    output = {}
    
    files = listdir(dirname)
    for f in files:
        output[f.split('.')[0].split('_')[-1]] = (pd.read_csv(dirname + f, index_col=0))
        
    return output

def getNormVarToNeighs(gamedata, m, n):
    # input is gamedata with all the runs
    # output is a dataframe with ['M', 'N', 'alpha', 'beta', 'meanNormVar']
    data = gamedata[(gamedata.M == m) & (gamedata.N == n)]
    outData = pd.DataFrame(columns=['neighs', 'M', 'N', 'alpha', 'beta', 'meanNormVar', 'var'])
    # for unique values in neighs column
    for neigh in pd.Series(data.neighs.values).unique():
        outData.loc[len(outData)] = [neigh, m, n, float(np.power(2,m))/n, float(neigh)/n, 
                                     data[data.neighs == neigh].mean()['normVar'], data[data.neighs == neigh].mean()['variance']]

    return outData.sort(columns='beta')

def plotNormVarToNeighs(games, mCondition, nConditions=[], plotFullGame=False):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(1, 1, 1)
    cols = itertools.cycle(colors.cnames)
    for k in games.keys():   
        n = games[k].N[0]
        if (n not in nConditions):
            del games[k]
        else:
            # reduce to only one m value
            games[k] = games[k][games[k].M == mCondition]        
            data = getNormVarToNeighs(games[k], mCondition, n)
            if not plotFullGame:
                data = data[data.beta != 1]
            plt.loglog(data.beta, data.meanNormVar, c=next(cols), label = "Network: {}".format(k), linewidth=2)
            plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    return games
```

```python
dirname = '../data/results/'
# g is a directory of full gamedata
games = readFullGamedata(dirname)

mCondition = 2
nConditions = [401, 403]
asd = plotNormVarToNeighs(games, mCondition, nConditions)
# print asd['gamedata_N_403_rounds_10000_vonNeumann.csv']

# for k in games.keys():
#     gSplit = k.split('_')
#     N = int(gSplit[gSplit.index('N') +1])
#     nVarNeighs = getNormVarToNeighs(games[k], mConditions, N)

# h = getNormVarToNeighs(g['gamedata_N_101_rounds_10000_patch.csv'], 2, 101)
```
