```python
# import neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
%matplotlib inline
```

```python
# various functions to read data from CSV
def readGameData(dirname, fileConditions):
    # inputs are 
    # dirname - string containing the folder name to read data from 
    # fileConditions - a list of conditions containing parameters like
    #                  memory, number of agents, etc.
    # outputs are
    # gamedata - contains information about every game
    # attdata - contains attendance per round for all the games
    files = listdir(dirname)
    fileConditions.append('.game')
    gamedata = pd.DataFrame(columns=['M', 'N', 'S', 'rounds', 'run', 'alpha', 'attendance mean', 'variance', 'normVar', 'stdDev'])
    attdata = {}
    for f in files:
        if all(con in f for con in fileConditions):
            filesplit = f.split('_')
            # for every file from data folder read it, then extract neccessary values
            data = pd.read_csv(dirname + f, index_col=0)
            gamedata.loc[len(gamedata)] = [int(filesplit[1]), int(filesplit[3]), int(filesplit[5]), 
                                           int(filesplit[7]), int(filesplit[9]), 
                                           (2**float(filesplit[1])/int(filesplit[3])), data.sum().mean(), 
                                           # variance, normalized variance
                                           data.sum().var(), (data.sum().var() / int(filesplit[3])),
                                           # standard deviation
                                           np.std(data.sum())]
            attdata[f] = data.sum()

    attdata = pd.DataFrame(attdata)
    # get rid of NaN instances caused by a game of only speculators not playing
    attdata.fillna(0, inplace=True)
    
    historyData = calculateHistory(attdata)
    
    fileConditions.remove('.game')
    return (gamedata, attdata, historyData)

def readScoreData(dirname, fileConditions):
    # inputs are 
    # dirname - string containing the folder name to read data from 
    # fileConditions - a list of conditions containing parameters like
    #                  memory, number of agents, etc.
    # outputs are
    # scoreData - contains score evolution by time for every agent of the game
    files = listdir(dirname)
    fileConditions.append('.score')
    scoreData = {}
    for f in files:
        if all(con in f for con in fileConditions):
            filesplit = f.split('_')
            # for every file from data folder that is score create a DataFrame
            # and put it in scoreData dictionary with filename as key
            scoreData[f] = pd.read_csv(dirname + f, index_col=0)
    
    fileConditions.remove('.score')        
    return scoreData

def readFullGameData(dirname, fileConditions, numberOfGames=1):
    # inputs are 
    # dirname - string containing the folder name to read data from 
    # fileConditions - a list of conditions containing parameters like
    #                  memory, number of agents, etc.
    # numberOfGames - number of games to read
    # outputs are
    # fullGameData - contains decisions by rounds for every agent of the game
    files = listdir(dirname)
    fileConditions.append('.game')
    fullGameData = {}
    filesToRead = []
    for f in files:
        if all(con in f for con in fileConditions):
            filesToRead.append(f)
            
    limit = np.minimum(numberOfGames, len(filesToRead))
    for i in range(limit):
        # for every file from data folder that is score create a DataFrame
        # and put it in fullGameData dictionary with filename as key
        fullGameData[filesToRead[i]] = pd.read_csv(dirname + filesToRead[i], index_col=0)

    fileConditions.remove('.game')        
    return fullGameData
    
# functions needed to study various behaviours
def calculateHistory(attdata):
    histori = {}
    for k in attdata.keys():
        filesplit = k.split('_')
        n = int(filesplit[3])
        histori[k] = [1 if x<=n/2 else 0 for x in attdata[k]] 
        
    histori = pd.DataFrame(histori)
    return histori

# plot attendance per rounds for various number of agents
def plotAtt(data, conditions, lowerLimit=0, upperLimit=-1):
    # parse the input (divide number of agents from rest of the conditions
    # that are true for every number of agents)
    nAgentsConditions = []
    otherConditions = []
    for a in conditions:
        if 'N_' in a:
            nAgentsConditions.append(a)
        else:
            otherConditions.append(a)
            
    print otherConditions
    for x in data.columns:
        if (all(j in x for j in otherConditions) and any(k in x for k in nAgentsConditions)):
            ax = data[x][lowerLimit:upperLimit].plot(legend=True)
            # set y axis limit from 0 to max Attendance 
#             ax.set_ylim(0, int(x.split('_')[3]))
            
    plt.legend(loc='lower left')
    
# plot Alpha to normalize variance for all the runs
def plotAlphaToNormVar(gamedata):
    alphaToNormVar = {}
    for x in range(len(gamedata)):
        if gamedata.alpha[x] in alphaToNormVar:
            alphaToNormVar[gamedata.alpha[x]].append(gamedata.normVar[x])
        else:
            alphaToNormVar[gamedata.alpha[x]] = [gamedata.normVar[x]]

    for x in alphaToNormVar.keys():
        alphaToNormVar[x] = sum(alphaToNormVar[x]) / len(alphaToNormVar[x])

    alphaToNormVar = pd.DataFrame({ 'alpha' : alphaToNormVar.keys(), 'normVar' : alphaToNormVar.values() })
    alphaToNormVar.plot(kind='scatter', x='alpha', y='normVar', loglog=True, legend=True)
    
def plotHistoryToProb(histori, k, conditions):
    histDict = {}
    for key in histori.keys():
        if all(j in key for j in conditions):
            for x in range(k, len(histori[key])-1):
                ckey = ''.join(str(a) for a in histori[key][x-k:x])
                ckey = str(int(ckey,2))
                if ckey not in histDict.keys():
                    histDict[ckey] = []
                histDict[ckey].append(histori[key][x])

    # after we have calculated the dictionary with
    # (key, value) = (history, decision list)
    # we have to calculate the probability about the decision
    # given a particular history
    if histDict:
        johnny = pd.DataFrame(columns=['memory', 'prob'])
        for j in histDict.keys():
            johnny.loc[len(johnny)] = [int(j), float(sum(histDict[j])) / len(histDict[j])]

        johnny.sort(columns='memory', inplace=True)
        johnny.plot(kind='bar', x='memory', y='prob')  
    
# plot memory to standard deviation of attendance for every game
def plotMemoryToStdDev(gamedata):
    gamedata.plot(kind='scatter', x='M', y='stdDev')
    
# plot the scores to the rounds
# we have to pass the scores pandas Dataframe and the number of agents we want to plot
def plotScores(scoreData, fileConditions, classic=0, producer=0, speculator=0, whatever=0, lowerRange=0, upperRange=-1):
    for k in scoreData.keys():
        if all(j in k for j in fileConditions):
            print k
            plotSingleScore(scoreData[k], classic, producer, speculator, whatever, lowerRange, upperRange)
    
def plotSingleScore(scores, classic=0, producer=0, speculator=0, whatever=0, lowerRange=0, upperRange=-1):
    cAgents = []
    pAgents = []
    sAgents = []
    wAgents = []
    for col in scores.columns:
        if 'classic' in col: cAgents.append(col)
        elif 'producer' in col: pAgents.append(col)
        elif 'speculator' in col: sAgents.append(col)
        else: wAgents.append(col)
      
    plt.figure()
    if (classic>0 and cAgents): 
        plt.plot(scores[cAgents[:classic]][lowerRange:upperRange].index, scores[cAgents[:classic]][lowerRange:upperRange], 'b')
    if (producer>0 and pAgents): 
        plt.plot(scores[pAgents[:producer]][lowerRange:upperRange].index, scores[pAgents[:producer]][lowerRange:upperRange], 'g')
    if (speculator>0 and sAgents): 
        plt.plot(scores[sAgents[:speculator]][lowerRange:upperRange].index, scores[sAgents[:speculator]][lowerRange:upperRange],'r')
    if (whatever>0 and wAgents): 
        plt.plot(scores[wAgents[:whatever]][lowerRange:upperRange].index, scores[wAgents[:whatever]][lowerRange:upperRange],'y')
        
        
#ACTIVE SPECULATORS PART
def calculateActiveSpeculators(fullGameData):
    activeSpeculators = {}
    for k in fullGameData.keys():
        AS = []
        ksplit = k.split('_')
        nOfProducers = int(ksplit[ksplit.index('prod') +1])
        nOfSpeculators = int(ksplit[ksplit.index('spec') +1])
        for j in xrange(fullGameData[k].shape[1]):
            AS.append(countNonZeros(fullGameData[k].ix[nOfProducers:,j]))
        activeSpeculators[k] = AS
    
    return activeSpeculators

def plotActiveSpeculators(data, numberOfGames=1, lowerlimit=0, upperlimit=-1):
    # inputs:
    # data - dictionary of Dataframes with full game data (all agents decisions round by round)
    # numberOfGames - number of games to plot
    # lowerlimit - lower limit for plotting
    # upperlimit - ...
    keys = data.keys()
    limit = numberOfGames if numberOfGames<len(keys) else len(keys) 
    plt.figure()
    for x in range(limit):
        ksplit = keys[x].split('_')
        nOfProducers = int(ksplit[ksplit.index('prod') +1])
        nOfSpeculators = int(ksplit[ksplit.index('spec') +1])
        if upperlimit<0 or upperlimit>len(data[keys[x]]):
            upperlimit = len(data[keys[x]])
        plt.plot(range(lowerlimit, upperlimit), data[keys[x]][lowerlimit:upperlimit])
        plt.ylim(0,nOfSpeculators)
    
def plotPriceClassic(attData, lam, conditions, lowerlimit=0, upperlimit=-1):
    price = {}
    keys = attData.keys()
    for k in keys:
        if all(con in k for con in conditions):
            price[k] = range(len(attData[k]))
            price[k][0] = 100
            ksplit = k.split('_')
            N = int(ksplit[ksplit.index('N') +1])
#             N = int(ksplit[ksplit.index('prod') +1]) + int(ksplit[ksplit.index('spec') +1])
            for x in xrange(1,len(attData[k])):
                price[k][x] = price[k][x-1] * np.exp( float(attData[k][x] - N/2) / lam )
                
    if price:
        plt.figure()
        for k in price.keys():
            if upperlimit<0 or upperlimit>len(price[k]):
                upperlimit = len(price[k])
            plt.plot(range(lowerlimit, upperlimit), price[k][lowerlimit:upperlimit])
        
    return price    

def countNonZeros(l):
    return np.invert(np.isnan(l)).sum()
```

```python
def plotPredictabilityToSpeculators(historyData, attData, conditions, brainSize):
    brainSizeConditions = []
    otherConditions = []
    for a in conditions:
        if 'M_' in a:
            brainSizeConditions.append(a)
        else:
            otherConditions.append(a)
            
    keys = []
    for k in historyData.keys():
        if all(j in k for j in otherConditions) and any(j in k for j in brainSizeConditions):
            keys.append(k)
            
    predDict = calculatePredictability(historyData, attData, keys, brainSize)
    # after we have calculated the dictionary with
    # (key, value) = (history, decision list)
    # we have to calculate the probability about the decision
    # given a particular history
    if predDict:
        print predDict
        
def calculatePredictability(historyData, attData, keys, brainSize):
    predDict = {}
    for k in keys:
        pred = {}
        ksplit = k.split('_')
        N = int(ksplit[ksplit.index('N') +1])
        for x in xrange(brainSize, len(historyData[k])-1):
            ckey = ''.join(str(a) for a in historyData[k][x-brainSize:x])
            ckey = str(int(ckey,2))
            if ckey not in pred.keys():
                pred[ckey] = []
            pred[ckey].append(historyData[k][x])

        for nk in pred.keys():
            pred[nk] = np.mean(pred[nk])
        
        predDict[k] = (1.0/2**brainSize) * np.square(pred.values()).sum()
        
    return predDict
    
fileConditions = ['M_2', 'M_3', 'M_4', 'M_5', 'M_6', 'M_7', 'M_8', 'M_9', 'N_101', 'run_11']
asd = plotPredictabilityToSpeculators(historyData, attData, fileConditions, 12)
```
